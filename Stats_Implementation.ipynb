```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Normal distribution dist >>> majority of the data ppoint is cluster around mean

mu,sigma =0,0.1
s=np.random .normal(mu,sigma ,size=1000)
print("Count of dps",len(s))
s
```

```python
plt.hist(s,30,density=True)
plt.show
```

```python
plt.hist(s,30)
plt.show
```

```python
sns.distplot(s)
```

```python
#Log normal dist >> whose logarithm is normally distributed>>right skewed
```

```python
l=np.random.lognormal(mean = 0,sigma=0.5,size=1000)
l=l[np.where(abs(l)<5)]
```

```python
plt.hist(l,bins=1000)
plt.show()
```

```python
#Log Normal ---> normal 
sns.distplot(np.log(l))
```

```python
sns.distplot(np.sqrt(l))
```

```python
#Chi-Square Distribution
```

```python
c=np.random.chisquare(df=1,size=1000)
c=c[np.where(abs(c)<10)]
plt.hist(c,bins=1000)
plt.show()
```

```python
c=np.random.chisquare(df=1,size=1000)
c=c[np.where(abs(c)<10)]
plt.hist(c,bins=1000)
plt.show()
```

```python
#chi square distribution with D.O.F=2
c=np.random.chisquare(df=2,size=1000)
c=c[np.where(abs(c)<10)]
plt.hist(c,bins=1000)
plt.show()
```

```python
#t-dISTRIBUTION with df 10 >>> tail is thinner/smaller #approx to normal distribution
```

```python
t=np.random.standard_t(df=10,size=1000)
t=t[np.where(abs(t)<10)]
plt.hist(t,bins=1000)
plt.show()
```

```python
#t-dISTRIBUTION with df 1 >>> tail is thicker/larger
t=np.random.standard_t(df=1,size=1000)
t=t[np.where(abs(t)<10)]
plt.hist(t,bins=1000)
plt.show()
```

```python
#F distribution
#dfnum is df for numerator, dfden is df for denominator
f=np.random.f(dfnum=1,dfden=1,size=1000)
f= f[np.where(abs(f)<10)]
plt.hist(f,bins=1000)
plt.show()
```
