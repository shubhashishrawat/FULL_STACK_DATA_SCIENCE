```python
import seaborn as sns 
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#Scaling >> brings all the features on the same scale 
```

#Why Scaling ?...... to reduce the computation time(faster optimization) and better interpretation

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
#Standardisation
mean=np.mean(df['total_bill'])
```

```python
std=np.std(df['total_bill'])
```

```python
std_data=[]
for i in list(df['total_bill']):
    zscore= (i-mean)/std
    std_data.append(zscore)
```

```python
std_data
```

```python
#Scaling is optional because it does not change the distribution of data.
#Before Scaling
sns.distplot(df['total_bill'])
```

```python
#After Scaling
sns.distplot(std_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Test data is unseen data , so you cannot learn any parameter from test and thats why you only apply the learned parameters from Train Data(Transform)
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']]) #it will calculate mu and sigma
```

```python
scaler.transform(df[['total_bill']])
```

```python
scaler.fit_transform(df[['total_bill','tip']])
```
